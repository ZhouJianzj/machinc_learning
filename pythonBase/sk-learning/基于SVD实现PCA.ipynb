```python
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# 
digits = load_digits()
X = digits.data
y = digits.target

#  PCA 
pca = PCA(n_components=2,svd_solver='full')

# 
X_reduced = pca.fit_transform(X)

# 
plt.figure(figsize=(8, 6))
#X_reduced[:, 0]  X_reduced[:, 1]: xyX_reduced [:, 0][:, 1]
# c=y: yscatter3012scatter
# cmap='tab10': 'tab10'10
# edgecolor='k': 'k'
# s=50: 50
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=y, cmap='tab10', edgecolor='k', s=50)
plt.colorbar(label='digit label')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('PCA of Digits Dataset')
plt.show()

```
