```python
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# 加载手写数字数据集
digits = load_digits()
X = digits.data
y = digits.target

# 创建 PCA 模型，将数据降至二维
pca = PCA(n_components=2,svd_solver='full')

# 对数据进行降维
X_reduced = pca.fit_transform(X)

# 绘制降维后的数据
plt.figure(figsize=(8, 6))
#X_reduced[:, 0] 和 X_reduced[:, 1]: 这些是要绘制的数据点的x和y坐标。通常，X_reduced 是一个二维数组，其中每行代表一个数据点，而每列代表一个特征。[:, 0]表示取所有行的第一个特征，[:, 1]表示取所有行的第二个特征。
# c=y: 这指定了每个数据点的颜色。y是一个数组，其中包含每个数据点的类别或标签。scatter函数根据这些类别自动为每个点分配颜色。例如，如果有3个类别，它们可能分别是0、1和2，那么scatter将会用不同的颜色表示这些类别。
# cmap='tab10': 这是指定用于着色的颜色映射。在这种情况下，'tab10'是一种内置的颜色映射，它包含了10种不同的颜色。它将被用来为不同的类别分配颜色。
# edgecolor='k': 这是指定数据点的边缘颜色。在这里，'k'表示黑色。
# s=50: 这是指定数据点的大小。在这里，所有的数据点都将具有相同的大小，即50。
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=y, cmap='tab10', edgecolor='k', s=50)
plt.colorbar(label='digit label')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('PCA of Digits Dataset')
plt.show()

```
