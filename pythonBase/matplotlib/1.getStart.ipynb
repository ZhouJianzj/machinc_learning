```python
# oo风格的图表
import numpy as np

# 生成100个0-2等间隔的数
x = np.linspace(0, 20, 20)  # Sample data.
fig, ax = plt.subplots(figsize=(5, 2), layout="constrained");
# 绘制第一条线
ax.plot(x, x, label="one",marker= 'o')
# 绘制第二条线
ax.plot(x, x*8, label="two",marker= 'x')
# 绘制第三条线
ax.plot(x, x*14, label="three")
ax.set_xlabel("x")
ax.set_ylabel("y")
#设置x轴刻度
# ax.set_xticks([1, 2, 3, 4, 5], ['A', 'B', 'C', 'D', 'E'])
ax.legend()
```
